This is a tiny repo of experimental utilties to help students learn
about linear algebra and such with complex numbers. You can display a phasor
by creating a "PhasorPlotter" instance and then calling its "plot" method
with a phasor (complex number). In the cell below is an example illustrating how
it works. Three phasors are plotted along with a grid to help see the scaling.

```python
p = PhasorPlotter()
p.plot(2+3j)
p.plot(-2+1j)
p.plot(-3-2j)
p.grid()
```

Below is another example with only a single phasor.

```python
p = PhasorPlotter()
p.plot(1+1j)
p.grid()
```
